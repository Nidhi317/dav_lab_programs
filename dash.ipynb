**Student Performance Dashboard**

```python
# =====================================
# üéì Interactive Student Performance Dashboard (for Google Colab)
# =====================================

# Import required libraries
from dash import Dash, dcc, html, Input, Output
import plotly.express as px
import pandas as pd

# -----------------------------
# Sample Data
# -----------------------------
data = {
    'Student': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'Study_Hours': [5, 8, 2, 9, 6, 7, 4, 3, 8, 10],
    'Maths': [78, 85, 60, 95, 82, 88, 70, 65, 90, 92],
    'Science': [72, 89, 58, 91, 80, 86, 68, 63, 87, 90],
    'English': [75, 82, 62, 94, 79, 84, 69, 64, 88, 91]
}

df = pd.DataFrame(data)
subjects = ['Maths', 'Science', 'English']

# -----------------------------
# Initialize JupyterDash app
# -----------------------------
app = Dash(__name__)
app.title = "Student Performance Dashboard"

# -----------------------------
# Layout
# -----------------------------
app.layout = html.Div([
    html.H1("üéì Student Performance Dashboard", style={'textAlign': 'center'}),

    html.Div([
        html.Label("Select Subject:"),
        dcc.Dropdown(
            id='subject-dropdown',
            options=[{'label': sub, 'value': sub} for sub in subjects],
            value='Maths',
            clearable=False
        ),
    ], style={'width': '30%', 'display': 'inline-block', 'padding': '10px'}),

    html.Div([
        html.Label("Filter by Study Hours:"),
        dcc.Slider(
            id='study-slider',
            min=df['Study_Hours'].min(),
            max=df['Study_Hours'].max(),
            step=1,
            value=df['Study_Hours'].max(),
            marks={i: str(i) for i in range(df['Study_Hours'].min(), df['Study_Hours'].max() + 1)}
        ),
    ], style={'width': '60%', 'display': 'inline-block', 'padding': '10px'}),

    html.Br(),

    html.Div([
        dcc.Graph(id='bar-chart', style={'width': '48%', 'display': 'inline-block'}),
        dcc.Graph(id='scatter-plot', style={'width': '48%', 'display': 'inline-block'})
    ])
])

# -----------------------------
# Callbacks
# -----------------------------
@app.callback(
    [Output('bar-chart', 'figure'),
     Output('scatter-plot', 'figure')],
    [Input('subject-dropdown', 'value'),
     Input('study-slider', 'value')]
)
def update_graphs(selected_subject, max_study_hours):
    filtered_df = df[df['Study_Hours'] <= max_study_hours]

    # Bar Chart: Average score by gender
    bar_fig = px.bar(
        filtered_df.groupby('Gender', as_index=False)[selected_subject].mean(),
        x='Gender', y=selected_subject,
        color='Gender',
        title=f"Average {selected_subject} Score by Gender"
    )

    # Scatter Plot: Study hours vs selected subject
    scatter_fig = px.scatter(
        filtered_df,
        x='Study_Hours', y=selected_subject,
        color='Gender',
        size=selected_subject,
        hover_data=['Student'],
        title=f"{selected_subject} Score vs Study Hours"
    )

    return bar_fig, scatter_fig

# -----------------------------
# Run inside Colab
# -----------------------------
app.run(mode='inline')
```

**Sales Dashboard**

```python
# =====================================
# üíπ Interactive Sales Dashboard (for Google Colab)
# =====================================

# Import required libraries
from dash import dcc, html, Input, Output
import plotly.express as px
import pandas as pd

# -----------------------------
# Sample Sales Data
# -----------------------------
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Region': ['North', 'South', 'East', 'West'] * 3,
    'Sales': [20000, 15000, 18000, 22000, 24000, 21000, 25000, 23000, 26000, 28000, 27000, 30000],
    'Profit': [4000, 3500, 3200, 5000, 5200, 4800, 5600, 5000, 6000, 6400, 6300, 7000]
}
df = pd.DataFrame(data)

# -----------------------------
# Initialize app
# -----------------------------
app = Dash(__name__)
app.title = "Sales Dashboard"

# -----------------------------
# Layout
# -----------------------------
app.layout = html.Div([
    html.H1("üíπ Sales Dashboard", style={'textAlign': 'center'}),

    html.Div([
        html.Label("Select Region:"),
        dcc.Dropdown(
            id='region-dropdown',
            options=[{'label': r, 'value': r} for r in df['Region'].unique()],
            value='North',
            clearable=False
        )
    ], style={'width': '30%', 'display': 'inline-block', 'padding': '10px'}),

    html.Br(),

    html.Div([
        dcc.Graph(id='sales-trend', style={'width': '48%', 'display': 'inline-block'}),
        dcc.Graph(id='profit-bar', style={'width': '48%', 'display': 'inline-block'})
    ])
])

# -----------------------------
# Callbacks
# -----------------------------
@app.callback(
    [Output('sales-trend', 'figure'),
     Output('profit-bar', 'figure')],
    [Input('region-dropdown', 'value')]
)
def update_dashboard(selected_region):
    filtered = df[df['Region'] == selected_region]

    sales_fig = px.line(filtered, x='Month', y='Sales', title=f'Sales Trend - {selected_region}', markers=True)
    profit_fig = px.bar(filtered, x='Month', y='Profit', title=f'Profit by Month - {selected_region}', color='Profit')

    return sales_fig, profit_fig

# -----------------------------
# Run app inline
# -----------------------------
app.run(mode='inline')
```

**COVID-19 Dashboard**

```python
# =====================================
# ü¶† Interactive COVID-19 Dashboard (for Google Colab)
# =====================================

# Install required libraries
from dash import dcc, html, Input, Output
import plotly.express as px
import pandas as pd

# -----------------------------
# Sample COVID-19 Data
# -----------------------------
data = {
    'Country': ['India', 'USA', 'Brazil', 'Russia', 'UK', 'France', 'Germany', 'Japan', 'China', 'Australia'],
    'Confirmed': [45000000, 98000000, 68000000, 21000000, 25000000, 39000000, 38000000, 3300000, 120000, 14000000],
    'Recovered': [44000000, 96000000, 67000000, 20000000, 24000000, 38000000, 37000000, 3100000, 119000, 13900000],
    'Deaths': [530000, 1100000, 700000, 400000, 300000, 280000, 250000, 45000, 5000, 20000],
    'Vaccination_Rate': [0.75, 0.8, 0.7, 0.65, 0.78, 0.76, 0.8, 0.9, 0.95, 0.88]
}

df = pd.DataFrame(data)

# -----------------------------
# Initialize app
# -----------------------------
app = Dash(__name__)
app.title = "COVID-19 Dashboard"

# -----------------------------
# Layout
# -----------------------------
app.layout = html.Div([
    html.H1("ü¶† COVID-19 Dashboard", style={'textAlign': 'center'}),

    html.Div([
        html.Label("Select Metric:"),
        dcc.Dropdown(
            id='metric-dropdown',
            options=[
                {'label': 'Confirmed Cases', 'value': 'Confirmed'},
                {'label': 'Recovered Cases', 'value': 'Recovered'},
                {'label': 'Deaths', 'value': 'Deaths'},
                {'label': 'Vaccination Rate', 'value': 'Vaccination_Rate'}
            ],
            value='Confirmed',
            clearable=False
        )
    ], style={'width': '40%', 'display': 'inline-block', 'padding': '10px'}),

    html.Br(),

    html.Div([
        dcc.Graph(id='bar-chart', style={'width': '48%', 'display': 'inline-block'}),
        dcc.Graph(id='scatter-chart', style={'width': '48%', 'display': 'inline-block'})
    ])
])

# -----------------------------
# Callback
# -----------------------------
@app.callback(
    [Output('bar-chart', 'figure'),
     Output('scatter-chart', 'figure')],
    [Input('metric-dropdown', 'value')]
)
def update_dashboard(selected_metric):
    bar_fig = px.bar(
        df.sort_values(selected_metric, ascending=False),
        x='Country', y=selected_metric,
        color='Country',
        title=f'{selected_metric} by Country'
    )

    scatter_fig = px.scatter(
        df, x='Vaccination_Rate', y=selected_metric,
        size=selected_metric, color='Country',
        hover_name='Country',
        title=f'{selected_metric} vs Vaccination Rate'
    )

    return bar_fig, scatter_fig

# -----------------------------
# Run inline in Colab
# -----------------------------
app.run(mode='inline')
```

**Patient Vitals Dashboard**

```python
# =====================================
# üè• Interactive Patient Vitals Dashboard (for Google Colab)
# =====================================

# Import required libraries
from dash import dcc, html, Input, Output
import plotly.express as px
import pandas as pd

# -----------------------------
# Sample Patient Vitals Data
# -----------------------------
data = {
    'Patient_ID': [f'P{i}' for i in range(1, 11)],
    'Age': [25, 38, 45, 60, 52, 33, 47, 58, 40, 29],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'Heart Rate (bpm)': [72, 80, 65, 90, 75, 85, 70, 88, 76, 82],
    'Blood Pressure (mmHg)': [120, 130, 110, 145, 125, 135, 118, 140, 122, 128],
    'Temperature (¬∞C)': [36.6, 37.2, 36.8, 38.1, 37.0, 37.5, 36.7, 38.3, 37.1, 36.9],
    'Oxygen Level (%)': [98, 97, 99, 94, 96, 95, 98, 93, 97, 99]
}

df = pd.DataFrame(data)

# -----------------------------
# Initialize app
# -----------------------------
app = Dash(__name__)
app.title = "Patient Vitals Dashboard"

# -----------------------------
# Layout
# -----------------------------
app.layout = html.Div([
    html.H1("üè• Patient Vitals Dashboard", style={'textAlign': 'center'}),

    html.Div([
        html.Label("Select Vital Sign:"),
        dcc.Dropdown(
            id='vital-dropdown',
            options=[
                {'label': 'Heart Rate (bpm)', 'value': 'Heart Rate (bpm)'},
                {'label': 'Blood Pressure (mmHg)', 'value': 'Blood Pressure (mmHg)'},
                {'label': 'Temperature (¬∞C)', 'value': 'Temperature (¬∞C)'},
                {'label': 'Oxygen Level (%)', 'value': 'Oxygen Level (%)'}
            ],
            value='Heart Rate (bpm)',
            clearable=False
        )
    ], style={'width': '40%', 'display': 'inline-block', 'padding': '10px'}),

    html.Br(),

    html.Div([
        dcc.Graph(id='bar-chart', style={'width': '48%', 'display': 'inline-block'}),
        dcc.Graph(id='scatter-plot', style={'width': '48%', 'display': 'inline-block'})
    ])
])

# -----------------------------
# Callbacks
# -----------------------------
@app.callback(
    [Output('bar-chart', 'figure'),
     Output('scatter-plot', 'figure')],
    [Input('vital-dropdown', 'value')]
)
def update_dashboard(selected_vital):
    bar_fig = px.bar(
        df,
        x='Patient_ID', y=selected_vital,
        color='Gender',
        title=f'{selected_vital} by Patient'
    )

    scatter_fig = px.scatter(
        df,
        x='Age', y=selected_vital,
        color='Gender',
        size='Oxygen Level (%)',
        hover_name='Patient_ID',
        title=f'{selected_vital} vs Age (Bubble size: Oxygen Level)'
    )

    return bar_fig, scatter_fig

# -----------------------------
# Run inline in Colab
# -----------------------------
app.run(mode='inline')
```
