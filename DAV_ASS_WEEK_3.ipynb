```python
import matplotlib.pyplot as plt

categories = ["Electronics", "Clothing", "Home Appliances", "Groceries"]
units_sold = [1200, 800, 350, 2600]

plt.bar(categories, units_sold, color="skyblue", edgecolor="black")
plt.title("Units Sold per Product Category (Monthly)")
plt.xlabel("Product Category")
plt.ylabel("Units Sold")
plt.show()

max_units = max(units_sold)
min_units = min(units_sold)

highest_category = categories[units_sold.index(max_units)]
lowest_category = categories[units_sold.index(min_units)]

print("Highest sales:", highest_category, "-", max_units, "units")
print("Lowest sales:", lowest_category, "-", min_units, "units")
```
