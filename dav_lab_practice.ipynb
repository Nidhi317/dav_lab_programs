**Merging**

```python
import pandas as pd

# Sample DataFrames
df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'ID': [2, 3, 4], 'Age': [25, 30, 35]})

# Merging with different join types
inner_join = pd.merge(df1, df2, on='ID', how='inner')  # Inner Join
outer_join = pd.merge(df1, df2, on='ID', how='outer')  # Outer Join
left_join = pd.merge(df1, df2, on='ID', how='left')    # Left Join
right_join = pd.merge(df1, df2, on='ID', how='right')  # Right Join

# Display results
print("Inner Join:\n", inner_join)
print("\nOuter Join:\n", outer_join)
print("\nLeft Join:\n", left_join)
print("\nRight Join:\n", right_join)
```

**Join**

```python
import pandas as pd

# Create two DataFrames
df1 = pd.DataFrame({'key': ['K0', 'K1', 'K2'], 'A': ['A0', 'A1', 'A2']})
df2 = pd.DataFrame({'key': ['K0', 'K1', 'K2'], 'B': ['B0', 'B1', 'B2']})

# Join DataFrames using their indexes
result = df1.set_index('key').join(df2.set_index('key'))
print(result)
```

**Row Wise Concatenation**

```python
import pandas as pd

# Example DataFrames
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})

# Concatenating along axis=0
result = pd.concat([df1, df2], axis=0)
print(result)
```

**Column Wise Concatenation**

```python
# Example DataFrames
df1 = pd.DataFrame({'A': [1, 2]}, index=[0, 1])
df2 = pd.DataFrame({'B': [3, 4]}, index=[0, 1])

# Concatenating along axis=1
result = pd.concat([df1, df2], axis=1)
print(result)
```

**Pivot Table**

```python
import pandas as pd

# Sample data
data = {
    'City': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Year': [2020, 2020, 2021, 2021, 2022, 2022],
    'Sales': [200, 150, 300, 250, 400, 350]
}

df = pd.DataFrame(data)

# Create pivot table
pivot = pd.pivot_table(df, values='Sales', index='Year', columns='City', aggfunc='sum')
print(pivot)
```

**Line Chart**

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, marker='o', color='pink', label='Line', linestyle = '--')
plt.scatter(x, y, color='red', label='Scatter')
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

**Bar Plot**

```python
import matplotlib.pyplot as plt

x = ['A', 'B', 'C', 'D']
y = [5, 7, 3, 8]

plt.bar(x, y, color='green', label='Bars', width=0.5)
plt.title("Bar Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.legend()
plt.show()
```

**Pie Chart**

```python
labels = ['Python', 'Java', 'C++', 'Ruby']
sizes = [40, 30, 20, 10]
colors = ['gold', 'lightblue', 'lightgreen', 'pink']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart")
plt.show()
```
